```python
%matplotlib notebook
import tessreduce as tr
import matplotlib.pyplot as plt
import numpy as np
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
ra = 189.1385817
dec = 11.2316535
#c = SkyCoord('10 18 13.12 +71 55 44.2', unit=(u.hourangle, u.deg), frame='icrs')
#ra = c.ra.deg
#dec = c.dec.deg
trc = tr.tessreduce(ra=ra,dec=dec)
```

```python
trc.reduce(diff_lc=True)
```

```python
trc.event_plotter()
```

```python
test = trc.Remove_background_var()
```

```python
plt.figure()
plt.plot(trc.lc[0],trc.lc[1],'.',label='standard lc')
plt.plot(test[0],test[1],'.',label='detrend lc')

plt.ylabel('Counts')
plt.xlabel('MJD')
plt.legend()
plt.savefig('detrend_comparison.png')
```
